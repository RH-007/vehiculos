```python
df = df_combinado.copy()

# normalizaciones básicas
df["url_low"]     = df["url_auto"].str.lower()
df["marca_slug"]  = df["marca"].str.lower().str.replace(r"\s+", "-", regex=True).str.replace(r"[^a-z0-9]+","-", regex=True)
df["modelo_slug"] = df["modelo"].str.lower().str.replace(r"\s+", "-", regex=True).str.replace(r"[^a-z0-9]+","-", regex=True)

# checks fila a fila (sin regex fancy)
df["brand_in_url"] = df.apply(lambda r: r["marca_slug"]  in r["url_low"], axis=1)
df["model_in_url"] = df.apply(lambda r: r["modelo_slug"] in r["url_low"], axis=1)
df["year_in_url"]  = df.apply(lambda r: str(r["año"]) in r["url_low"], axis=1)  # usa "anio" si tu col se llama así

df["url_ok"] = df[["brand_in_url","model_in_url","year_in_url"]].all(axis=1)

sospechosos = df.loc[~df["url_ok"], ["titulo", "tipo","marca","modelo","año", "precio","url_auto","brand_in_url","model_in_url","year_in_url"]]
sospechosos.head(20)
```

```python
enlaces = sospechosos["url_auto"].to_list()
titulo = sospechosos["titulo"].to_list()
tipo = sospechosos["tipo"].replace({"usados":"usado", "nuevos":"nuevo", "seminuevo":"seminuevo"}).str.lower().to_list()
marca = sospechosos["marca"].str.lower().to_list()
modelo = sospechosos["modelo"].str.lower().to_list() 
año = sospechosos["año"].to_list()

for enlace in enlaces:
    for titulo_, marca_, tipo_, modelo_, año_ in zip(titulo, marca,tipo, modelo, año):
        if marca_ in enlace and modelo_ in enlace and str(año_) in enlace and tipo_ in enlace:
            print(f"OK: {enlace} - {titulo_} - {tipo_} - {marca_} - {modelo_} - {año_}")
        # else:
        #     print(f"REVISAR: {enlace} - {titulo_} - {tipo_} - {marca_} - {modelo_} - {año_}")
```

```python
labels_precio = [ "menor a $5K", "$5K - $10K", "$10K - $20K", "$20K - $30K", "$30K - $40K", "$40K a más"]

df_combinado["precio_etiqueta"] = pd.cut(
    df_combinado["precio_fix"],
    bins=[-1, 5000, 10000, 20000, 30000, 40000, float('inf')],
    labels=labels_precio
    )
    
```
